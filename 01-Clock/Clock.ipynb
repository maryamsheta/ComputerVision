```python
import cv2
import math
import datetime
import numpy as np
```

```python
cv2.namedWindow("Clock")

main_color = (106, 53, 12)
hours_color = (255, 158, 39)
seconds_color = (238,238,238)
bg = (238,238,238)

degree = np.arange(300,660,30)
degree_seconds = np.arange(0,360,6)

while True:
    date = datetime.date.today()
    formatted_date = date.strftime("%B %d, %Y")
    time = datetime.datetime.now()
    formatted_time = time.strftime("%H:%M:%S")

    img = np.full((800,800,3),bg,np.uint8)
    
    cv2.putText(img,f"{formatted_date} {formatted_time}",(185,700),cv2.FONT_ITALIC,1,main_color,5)
    cv2.circle(img,(400,350),240,main_color,-1)

    for i in range(0,60):

        x = int(400+240*math.cos(math.radians(degree_seconds[i])))
        y = int(350+240*math.sin(math.radians(degree_seconds[i])))
        
        cv2.circle(img,(x,y),5,seconds_color,-1)


    for i in range(0,12):

        x_text = int(390+280*math.cos(math.radians(degree[i])))
        y_text = int(360+280*math.sin(math.radians(degree[i])))

        x_circle = int(400+240*math.cos(math.radians(degree[i])))
        y_circle = int(350+240*math.sin(math.radians(degree[i])))

        cv2.putText(img,str(i+1),(x_text,y_text),cv2.FONT_ITALIC,1,hours_color,5)
        cv2.circle(img,(x_circle,y_circle),8,hours_color,-1)

    hour = time.hour+time.minute/60+time.second/3600+time.microsecond/3600000000
    hours_angle = (hour/12*360)
    x_hour = int(400+100*math.sin(math.radians(hours_angle))) 
    y_hour = int(350-100*math.cos(math.radians(hours_angle)))  
    cv2.line(img,(400,350),(x_hour, y_hour),bg,3)

    minute  = time.minute+time.second/60+time.microsecond/60000000
    minutes_angle = minute/60*360
    x_minute = int(400+150*math.sin(math.radians(minutes_angle))) 
    y_minute = int(350-150*math.cos(math.radians(minutes_angle)))  
    cv2.line(img,(400,350),(x_minute, y_minute),bg,2)

    second = (time.second+time.microsecond/1000000)
    seconds_angle = second/60*360
    x_second = int(400+200*math.sin(math.radians(seconds_angle))) 
    y_second = int(350-200*math.cos(math.radians(seconds_angle)))  
    cv2.line(img,(400,350),(x_second, y_second),bg,1)

    cv2.circle(img,(400,350),8,hours_color,-1)
    
    cv2.imshow("Clock",img)


    k=cv2.waitKey(1)

    if k==ord("q"):
        break

cv2.destroyAllWindows()
```
