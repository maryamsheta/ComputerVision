```python
import cv2
import random
import numpy as np
```

### Restarts Automatically When Losing

```python
cv2.namedWindow("Snake")

window_x = 500
window_y = 500

snake_x = window_x//2
snake_y = window_y//2

snake_points = [[snake_x,snake_y]]

food = random.randint(50,window_x-50)

direction = (1,0)

length = 10

while True: 

    img = np.full((window_x,window_y,3),(238,238,238),np.uint8)

    cv2.circle(img,(food,food),5,(0,0,255),-1)

    for point in snake_points:
        cv2.circle(img,(point[0],point[1]),3,(106, 53, 12),-1)
        
    if snake_x in range(food-5,food+6) and snake_y in range(food-5,food+6):

        food = random.randint(50,window_x-50)

        while [food,food] in snake_points:
            food = random.randint(50,window_x-50)

        cv2.circle(img,(food,food),5,(0,0,255),-1)
        length+=25

    if snake_x > window_x-5 or snake_y > window_y-5 or snake_x == 5 or snake_y == 5 or [snake_x,snake_y] in snake_points[1:]:
        snake_x = window_x//2
        snake_y = window_y//2
        snake_points = [[snake_x,snake_y]]
        food = random.randint(50,window_x-50)
        direction = (1,0)
        length = 10
            
    cv2.imshow("Snake",img)

    k=cv2.waitKey(1)

    if k == ord("q"):
        break
    elif k == ord("w") and direction != (0,1): 
        direction = (0,-1)
    elif k == ord("s") and direction != (0,-1): 
        direction = (0,1)
    elif k == ord("a") and direction != (1,0):
        direction = (-1,0)
    elif k == ord("d") and direction != (-1,0):
        direction = (1,0)

    snake_x+=direction[0]
    snake_y+=direction[1]
    snake_points.insert(0,[snake_x,snake_y])
    if len(snake_points) > length:
        snake_points.pop()

cv2.destroyAllWindows()
```

### End Screen (GameOver, Score & Restart)

```python
cv2.namedWindow("Snake")

window_x = 500
window_y = 500

snake_x = window_x//2
snake_y = window_y//2

snake_points = [[snake_x,snake_y]]

food = random.randint(50,window_x-50)

direction = (1,0)

length = 10

score = 0

playing = True

while True: 

    img = np.full((window_x,window_y,3),(238,238,238),np.uint8)

    cv2.circle(img,(food,food),5,(0,0,255),-1)

    for point in snake_points:
        cv2.circle(img,(point[0],point[1]),3,(106, 53, 12),-1)
        
    if snake_x in range(food-5,food+6) and snake_y in range(food-5,food+6):

        food = random.randint(50,window_x-50)

        while [food,food] in snake_points:
            food = random.randint(50,window_x-50)

        cv2.circle(img,(food,food),5,(0,0,255),-1)
        length+=25
        score+=1

    if snake_x > window_x-5 or snake_y > window_y-5 or snake_x < 5 or snake_y < 5 or [snake_x,snake_y] in snake_points[1:]:
        cv2.putText(img,"GAME OVER!",(window_x//2-90,window_y//2),cv2.FONT_ITALIC,1,(0,0,255),5)
        cv2.putText(img,f"Score: {score}",(window_x//2-60,window_y//2+30),cv2.FONT_ITALIC,1,(0,0,255),3)
        cv2.putText(img,"Press r To Restart",(window_x//2-70,window_y//2+50),cv2.FONT_ITALIC,.5,(0,0,255),2)
        cv2.imshow("Snake", img)
        playing = False
            
    cv2.imshow("Snake",img)

    k=cv2.waitKey(1)

    if k == ord("q"):
        break
    elif k == ord("w") and direction != (0,1) and playing: 
        direction = (0,-1)
    elif k == ord("s") and direction != (0,-1) and playing: 
        direction = (0,1)
    elif k == ord("a") and direction != (1,0) and playing:
        direction = (-1,0)
    elif k == ord("d") and direction != (-1,0) and playing:
        direction = (1,0)
    elif k==ord("r"):
        playing = True
        snake_x = window_x//2
        snake_y = window_y//2
        snake_points = [[snake_x,snake_y]]
        food = random.randint(50,window_x-50)
        direction = (1,0)
        length = 10
        score = 0

    if playing: 
        snake_x+=direction[0]
        snake_y+=direction[1]
        snake_points.insert(0,[snake_x,snake_y])
        if len(snake_points) > length:
            snake_points.pop()

cv2.destroyAllWindows()
```
